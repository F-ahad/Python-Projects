# BASIC PROJECT OF PYTHON MATPLOTLIB MODULE

### THIS DOCUMENT CONTAINS THE BASIC PROGRAMS OF PYTHON'S MATPLOTLIB MODULE

### PLOT GRAPH

```python
import matplotlib.pyplot as plt

x=[1,3,5,7]
y=[2,6,8,9]
clr=["b", "r", "y", "g"]
plt.plot(x,y, color="black",linestyle=":", linewidth="2")
plt.title("Example Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

### BAR PLOT

```python
import matplotlib.pyplot as plt
import numpy as np

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [85,65,76,89,87]
z= [60,54,61,76,79]

width=0.3
i=np.arange(len(x))
ii=[j+width for j in i] 

plt.xlabel("Languages",fontsize=12, fontname='Times New Roman')
plt.ylabel("Popularity/useage", fontsize=12, fontname='Times New Roman')
plt.title("Language Trend", fontsize=15, fontweight='bold', fontname="Times New Roman",y=1.05)

plt.bar(i,y,width, color="pink", label="Popularity")
plt.bar(ii,z,width,color="maroon", label="Useage")

plt.xticks(i+width/2,x, rotation=45)
plt.legend()
plt.show()
```

### SCATTER PLOT

```python
import matplotlib.pyplot as plt

x= x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [85,65,76,89,87]
z= [60,54,61,76,79]

sizes1=[100,200,400,600,300]
sizes2=[300,600,400,200,100]
color1=[10,25,40, 90, 100]
color2=["r", "m", "g", "y", "y"]

plt.xlabel("Languages",fontsize=12, fontname='Times New Roman')
plt.ylabel("Popularity/useage", fontsize=12, fontname='Times New Roman')
plt.title("Language Trend", fontsize=15, fontweight='bold', fontname="Times New Roman")
    
scatter=plt.scatter(x,y, c=color1,cmap="viridis",s=sizes1)
plt.scatter(x,z,c=color2 ,s=sizes2)

color_bar=plt.colorbar(scatter)
color_bar.set_label("Color Scale")

plt.show()
```

### PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]

colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lavender']
explode_space=[0.2,0.0,0.0,0.0,0.0]

plt.pie(y, explode=explode_space
        , labels=x
        , shadow=True
        , colors=colors
        , autopct="%d%%"
        , radius=1.2
        , labeldistance=1.15
        , textprops={"fontname":"Times New Roman"}
        ,wedgeprops={"linewidth":5, "edgecolor":"black"})
plt.title("Golbal Useage of Programming Languages", fontname="Times New Roman", y=1.1)
plt.legend(loc=5)
plt.show()
```

### DONUT/RING PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]
explode_space=[0.2,0.0,0.0,0.0,0.00]

plt.pie(y, explode=explode_space, labels=x, radius=1.2)
plt.pie([1], colors="w", radius=0.8)
plt.title("Golbal Useage of Programming Languages", fontname="Times New Roman", y=1.1)

plt.show()
```

### ANOTHER WAY TO MAKE DONUT/RING PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]

plt.pie(y, labels=x, radius=1.2)
plt.title("Golbal Useage of Programming Languages", fontname="Times New Roman", y=1.1)

circle=plt.Circle(xy=(0,0), radius=0.8, facecolor="w")
plt.gca().add_artist(circle)
plt.show()
```
